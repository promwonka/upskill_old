```python
df1=bm[['Goals','Started','Home','Away','Right Foot','Left Foot','Header','Cross','Penalty','Open Play','Assists','Minutes per Yellow Cards']]
df1
```

```python
def minmax(col):
    ncol=[]
    mx=col.max()
    mn=col.min()
    for i in col:
        ncol.append((i-mn)/(mx-mn))
    return ncol
```

```python
dfn1=df1.apply(minmax,axis=0)
```

```python
dfn_sum1=dfn1.sum(axis=1).sort_values(ascending=False)
dfn_sum1
```
