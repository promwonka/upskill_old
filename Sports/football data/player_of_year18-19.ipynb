```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("C:/Users/Adith/Desktop/football_analysis/Football/Player Stats 2018-19.csv")
```

```python
new_data.dtypes
```

```python
home=new_data['Home'].astype(str).astype(int)
Away=new_data['Away'].astype(str).astype(int)
rft=new_data['Right Foot'].astype(str).astype(int)
lft=new_data['Left Foot'].astype(str).astype(int)
hdr=new_data['Header'].astype(str).astype(int)
dfk=new_data['Direct Free Kick'].astype(str).astype(int)
myc=new_data['Minutes per Yellow Cards'].astype(str).astype(float)
op=new_data['Open Play'].astype(str).astype(int)
cross=new_data['Cross'].astype(str).astype(int)
penalty=new_data['Penalty'].astype(str).astype(int)
```

```python
new_data = new_data.drop(['Home','Away','Right Foot','Left Foot','Header',
         'Direct Free Kick','Open Play','Cross','Penalty'], axis=1)
```

```python
new_data["Home"]=home
new_data['Away']=Away
new_data['Right Foot']=rft
new_data['Left Foot']=lft
new_data['Header']=hdr
new_data['Direct Free Kick']=dfk
new_data['Minutes per Yellow Cards']=myc
new_data['Open Play']=op
new_data['Cross']=cross
new_data['Penalty']=penalty
```

```python
df1 = df1.drop(['Home','Away','Right Foot','Left Foot','Header',
         'Direct Free Kick','Open Play','Cross','Penalty'], axis=1)
```

```python
df1["Home"]=home
df1['Away']=Away
df1['Right Foot']=rft
df1['Left Foot']=lft
df1['Header']=hdr
df1['Direct Free Kick']=dfk
df1['Minutes per Yellow Cards']=myc
df1['Open Play']=op
df1['Cross']=cross
df1['Penalty']=penalty
```

```python
df1
```

```python
def minmax(col):
    ncol=[]
    mx=col.max()
    mn=col.min()
    for i in col:
        ncol.append((i-mn)/(mx-mn))
    return ncol
```

```python
dfn1=df1.apply(minmax,axis=0)
```

```python
dfn1
```

```python
dfn_sum1=dfn1.sum(axis=1).sort_values(ascending=False)
dfn_sum1
```
