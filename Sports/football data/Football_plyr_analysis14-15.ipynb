```python
home=df2['Home'].astype(str).astype(int)
Away=df2['Away'].astype(str).astype(int)
rft=df2['Right Foot'].astype(str).astype(int)
lft=df2['Left Foot'].astype(str).astype(int)
hdr=df2['Header'].astype(str).astype(int)
dfk=df2['Direct Free Kick'].astype(str).astype(int)
myc=df2['Minutes per Yellow Cards'].astype(str).astype(float)
op=df2['Open Play'].astype(str).astype(int)
cross=df2['Cross'].astype(str).astype(int)
penalty=df2['Penalty'].astype(str).astype(int)
```

```python
df2 = df2.drop(['Home','Away','Right Foot','Left Foot','Header',
         'Direct Free Kick','Open Play','Cross','Penalty'], axis=1)
```

```python
df2["Home"]=home
df2['Away']=Away
df2['Right Foot']=rft
df2['Left Foot']=lft
df2['Header']=hdr
df2['Direct Free Kick']=dfk
df2['Minutes per Yellow Cards']=myc
df2['Open Play']=op
df2['Cross']=cross
df2['Penalty']=penalty
```

```python
df2
```

```python
def minmax(col):
    ncol=[]
    mx=col.max()
    mn=col.min()
    for i in col:
        ncol.append((i-mn)/(mx-mn))
    return ncol
```

```python
dfn=df2.apply(minmax,axis=0)
```

```python
dfn
```

```python
dfn_sum=dfn.sum(axis=1).sort_values(ascending=False)
dfn_sum
```

```python
ix1=dfn_sum.index[:10]
ix1
```
