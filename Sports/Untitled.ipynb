```python
import pandas as pd
```

```python
df = pd.read_csv('data_viz.csv')
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date']).dt.date
df['date'].max()
```

```python
df['week'] = pd.DatetimeIndex(df['date']).week
df['year'] = pd.DatetimeIndex(df['date']).year
df['QY']= df['year'].astype(str) + 'W' + df['week'].astype(str)
```

```python
df
```

```python
df.sort_values("QY", axis = 0, ascending = True, 
                 inplace = True, na_position ='last') 
qy_val = df.QY.unique() 
qy_val
```

```python
i = 0
while (i<l):
    
    globals()['df_passing%s' % i] = globals()['df%s' % i].groupby('name').passing.sum().reset_index()
    globals()['df_passing%s' % i].columns = ['name','passing']
    
    globals()['df_dribbling%s' % i] = globals()['df%s' % i].groupby('name').dribbling.sum().reset_index()
    globals()['df_dribbling%s' % i].columns = ['name','dribbling']
    
    globals()['df_kicking%s' % i] = globals()['df%s' % i].groupby('name').kicking.sum().reset_index()
    globals()['df_kicking%s' % i].columns = ['name','kicking']
    
    globals()['df_user%s' % i] = pd.merge(globals()['df_passing%s' % i], globals()['df_dribbling%s' % i], on='name')  
    globals()['df_user%s' % i] = pd.merge(globals()['df_user%s' % i], globals()['df_kicking%s' % i], on='name')
    
    #1
    
    globals()['df_weight_pass%s' % i] = globals()['df%s' % i].groupby('name')['weight of the pass'].sum().reset_index()
    globals()['df_weight_pass%s' % i].columns = ['name','weight of the pass']
    
    globals()['df_angle_kick%s' % i] = globals()['df%s' % i].groupby('name')['angle of kick (p)'].sum().reset_index()
    globals()['df_angle_kick%s' % i].columns = ['name','angle of kick (p)']
    
    globals()['df_pos_aware%s' % i] = globals()['df%s' % i].groupby('name')['position awareness'].sum().reset_index()
    globals()['df_pos_aware%s' % i].columns = ['name','position awareness']
    
    globals()['df_user_1%s' % i] = pd.merge(globals()['df_weight_pass%s' % i], globals()['df_angle_kick%s' % i], on='name')  
    globals()['df_user_1%s' % i] = pd.merge(globals()['df_user_1%s' % i], globals()['df_pos_aware%s' % i], on='name')
    
    #2
    
    globals()['df_the_cut%s' % i] = globals()['df%s' % i].groupby('name')['the cut'].sum().reset_index()
    globals()['df_the_cut%s' % i].columns = ['name','the cut']
    
    globals()['df_lunge%s' % i] = globals()['df%s' % i].groupby('name')['lunge'].sum().reset_index()
    globals()['df_lunge%s' % i].columns = ['name','lunge']
    
    globals()['df_scissors%s' % i] = globals()['df%s' % i].groupby('name')['scissors'].sum().reset_index()
    globals()['df_scissors%s' % i].columns = ['name','scissors']
    
    
    globals()['df_fake_kick%s' % i] = globals()['df%s' % i].groupby('name')['fake kick'].sum().reset_index()
    globals()['df_fake_kick%s' % i].columns = ['name',' fake kick']
    
    globals()['df_user_2%s' % i] = pd.merge(globals()['df_the_cut%s' % i], globals()['df_lunge%s' % i], on='name')  
    globals()['df_user_2%s' % i] = pd.merge(globals()['df_user_2%s' % i], globals()['df_scissors%s' % i], on='name')
    globals()['df_user_2%s' % i] = pd.merge(globals()['df_user_2%s' % i], globals()['df_fake_kick%s' % i], on='name')
    
    
    #3
    
    globals()['df_kick_str%s' % i] = globals()['df%s' % i].groupby('name')['kick strength'].sum().reset_index()
    globals()['df_kick_str%s' % i].columns = ['name','kick strength']
    
    globals()['df_accuracy%s' % i] = globals()['df%s' % i].groupby('name')['accuracy'].sum().reset_index()
    globals()['df_accuracy%s' % i].columns = ['name','accuracy']
    
    globals()['df_kick_angle%s' % i] = globals()['df%s' % i].groupby('name')['angle of kick'].sum().reset_index()
    globals()['df_kick_angle%s' % i].columns = ['name','angle of kick']
    
    
      
    globals()['df_user_3%s' % i] = pd.merge(globals()['df_kick_str%s' % i], globals()['df_accuracy%s' % i], on='name')  
    globals()['df_user_3%s' % i] = pd.merge(globals()['df_user_3%s' % i], globals()['df_kick_angle%s' % i], on='name')
    
    #all
    
    globals()['df_user%s' % i] = pd.merge(globals()['df_user%s' % i], globals()['df_user_1%s' % i], on='name')  
    globals()['df_user%s' % i] = pd.merge(globals()['df_user%s' % i], globals()['df_user_2%s' % i], on='name')
    globals()['df_user%s' % i] = pd.merge(globals()['df_user%s' % i], globals()['df_user_3%s' % i], on='name')
    
    
    globals()['df_user%s' % i]['QY'] = qy_val[i]
    
    i= i+1
```

```python
df_comb_final_0 = df_comb.groupby('name').passing.sum().reset_index()
```

```python
df_comb_final_1 = df_comb.groupby('name').dribbling.sum().reset_index()
```

```python
df_comb_final_2 = df_comb.groupby('name').kicking.sum().reset_index()
```

```python
df_comb_final = pd.merge(df_comb_final_0, df_comb_final_1, on='name')
```

```python
df_comb_final = pd.merge(df_comb_final, df_comb_final_2, on='name')
```

```python
df_comb_final
```

```python
df_comb_final['passing'] = df_comb_final.passing.rank(pct = True) 
df_comb_final['passing'] = df_comb_final['passing']*100

df_comb_final['dribbling'] = df_comb_final.dribbling.rank(pct = True) 
df_comb_final['dribbling'] = df_comb_final['dribbling']*100

df_comb_final['kicking'] = df_comb_final.kicking.rank(pct = True) 
df_comb_final['kicking'] = df_comb_final['kicking']*100

df_comb_final['overall_score'] = df_comb_final['passing'] + df_comb_final['dribbling'] + df_comb_final['kicking']
df_comb_final['overall_score'] = df_comb_final['overall_score']/3
```

```python
df_comb_final['Overall Segment'] = 'Need Attention'
df_comb_final.loc[df_comb_final['overall_score'] >= 25,'Overall Segment'] = 'Need Improvement' 
df_comb_final.loc[df_comb_final['overall_score'] >= 50,'Overall Segment'] = 'Good' 
df_comb_final.loc[df_comb_final['overall_score'] >= 75,'Overall Segment'] = 'Top' 
```

```python
df_comb_final['Passing Segment'] = 'Need Attention'
df_comb_final.loc[df_comb_final['passing'] >= 25,'Passing Segment'] = 'Need Improvement' 
df_comb_final.loc[df_comb_final['passing'] >= 50,'Passing Segment'] = 'Good' 
df_comb_final.loc[df_comb_final['passing'] >= 75,'Passing Segment'] = 'Top' 
```

```python
df_comb_final['Dribbling Segment'] = 'Need Attention'
df_comb_final.loc[df_comb_final['dribbling'] >= 25,'Dribbling Segment'] = 'Need Improvement' 
df_comb_final.loc[df_comb_final['dribbling'] >= 50,'Dribbling Segment'] = 'Good' 
df_comb_final.loc[df_comb_final['dribbling'] >= 75,'Dribbling Segment'] = 'Top' 
```

```python
df_comb_final['Kicking Segment'] = 'Need Attention'
df_comb_final.loc[df_comb_final['kicking'] >= 25,'Kicking Segment'] = 'Need Improvement' 
df_comb_final.loc[df_comb_final['kicking'] >= 50,'Kicking Segment'] = 'Good' 
df_comb_final.loc[df_comb_final['kicking'] >= 75,'Kicking Segment'] = 'Top' 
```
